```python
#scikit-learn
from sklearn.linear_model import LinearRegression
#es importante es separar los datos de entrenamiento de los datos y los datos de prueba para la validación
from sklearn.model_selection import train_test_split
#X mayuscula representa una matriz
X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml['Precio_Millon']

#Ahora hago el entrenamiento, test_size me dice que va entrenar con el 75% de los datos y va probar con el 25% , es importante definir el estado de aletoriedad
#por eso hay que pasar random_state, porque de esta manera, defino el estado de aletoriedad
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state = 99)
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X_train,y_train)
```

```python
y_predict_test= modelo.predict(X_test)
```

```python
#Cuanto el valor es mas cercano a 1, el modelo se acerca mejor
from sklearn.metrics import mean_absolute_error, r2_score

baseline_mae = mean_absolute_error(y_test, y_predict_test)
#baseline_mae

baseline_r2 = r2_score(y_test, y_predict_test)
baseline_r2
```

```python
#Despues de esto vemos que el codigo de predicción no esta performando bien, ahora es continuar con los experimentos, tomando tres variables mas


X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banios','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']] 

y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```

ojo, que para esta ultima corrida, se saco los outlier para mejorar el error medio y el R2, esto quiere decir que los datos se estan comportando mejor y tambipen se puede agregar más variables. También es importante ver la matriz de correlación
